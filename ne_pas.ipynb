```python
def find(my_path, my_ext, string_to_find, titolo_file):
    dirs = os.listdir(my_path)
    longer = len(string_to_find)
    longer_pas = 3
    totale = []
    path = "C:\Users\Oreste\Desktop\\" + titolo_file + ".txt"
    mon_fichier = open(path,'a') 
    string_to_find_pas = "pas"
    
    for file in dirs:
        count = 0
        quantita = 0
        if file[(len(file) - len(my_ext)):] == my_ext :
            print "\n"
            print(file)
            print "\n" 
            mon_fichier.write("\n")
            mon_fichier.write(file)
            mon_fichier.write("\n")
            with open(my_path + "\\" + file,'r') as file_in :
                for line in file_in:
                    x=line
                    count = count + 1
                    indice = x.find(string_to_find)
                    if (indice != -1 ) and (ord(x[indice+ longer])>122 or ord(x[indice+ longer])<65) and (ord(x[indice-1])>122 or ord(x[indice-1])<65) :
                        x = x[indice+longer:]
                        indice_pas = x.find(string_to_find_pas)
                    else:
                        indice_pas = -1
                    for i in range(len(x)):        
                        if (indice != -1 ) and (indice_pas != -1 ) and (ord(x[indice_pas+ longer_pas])>122 or ord(x[indice_pas+ longer_pas])<65) and (ord(x[indice_pas-1])>122 or ord(x[indice_pas-1])<65) :       
                            print " the line %d"%(count)
                            quantita += 1
                            print "ne" + x[:indice_pas+longer_pas]
                            print "\n"
                            mon_fichier.write(" the line %d"%(count))
                            mon_fichier.write(" :   ne"+x[:indice_pas+longer_pas])
                            mon_fichier.write("\n")
                            x = x[indice_pas+longer_pas:]
                            indice = x.find(string_to_find) 
                            if (indice != -1 ) and (ord(x[indice+ longer])>122 or ord(x[indice+ longer])<65) and (ord(x[indice-1])>122 or ord(x[indice-1])<65) :
                                x = x[indice+longer:]
                                indice_pas = x.find(string_to_find_pas)
                            else :
                                indice_pas = -1
            totale += [quantita]
            mon_fichier.write("\n")
            mon_fichier.write("quantita = %d"%(quantita))
            mon_fichier.write("\n")
            mon_fichier.write("\n")
    mon_fichier.close()
    return totale
```
