```python
def find(my_path, my_ext, titolo_file):
    dirs = os.listdir(my_path)
    longer = 3
    totale = []
    #path = "C:\Users\Oreste\Desktop\\" + titolo_file + ".txt"
    path = "C:\Users\IBM_ADMIN\Desktop\\" + titolo_file + ".txt"
    mon_fichier = open(path,'a') 
    string_to_find = "pas"
    
    for file in dirs:
        count = 0
        quantita = 0
        if file[(len(file) - len(my_ext)):] == my_ext :
            print "\n"
            print(file)
            print "\n" 
            mon_fichier.write("\n")
            mon_fichier.write(file)
            mon_fichier.write("\n")
            with open(my_path + "\\" + file,'r') as file_in :
                for line in file_in:
                    x=line
                    count = count + 1
                    indice = x.find(string_to_find)        
                    while (indice != -1 ) and (ord(x[indice+ longer])>122 or ord(x[indice+ longer])<65) and (ord(x[indice-1])>122 or ord(x[indice-1])<65) :   
                        if ((x.find("ne")>=indice) or (x.find("ne")==-1)) and ((x.find("n'")>=indice) or (x.find("n'")==-1)) and ((x.find("Ne")==-1) or (x.find("Ne")>=indice)) and ((x.find("N'")==-1) or (x.find("N'")>=indice)):
                            print " the line %d"%(count)
                            quantita += 1
                            if indice-200 > 0 :
                                print x[indice-200:indice+longer]
                            else :
                                print x[:indice+longer]  
                            print(x.find("n’"))
                            print "\n"
                            mon_fichier.write(" the line %d"%(count))
                            if indice-200 > 0 :
                                mon_fichier.write(x[indice-200:indice+longer])
                            else :
                                mon_fichier.write(x[:indice+longer])
                            mon_fichier.write("\n")
                        x = x[indice+longer:]
                        indice = x.find(string_to_find) 
                           
            totale += [quantita]
            mon_fichier.write("\n")
            mon_fichier.write("quantita = %d"%(quantita))
            mon_fichier.write("\n")
            mon_fichier.write("\n")
    mon_fichier.close()
    return totale
```
